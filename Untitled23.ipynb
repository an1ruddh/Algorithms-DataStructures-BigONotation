```python
import pandas as pd

# Load data from the first JSON file
with open('/content/class2_dataset.jsonl', 'r') as file:
    data1 = pd.read_json(file, lines=True)

# Load data from the second JSON file
with open('/content/class3_dataset.jsonl', 'r') as file:
    data2 = pd.read_json(file, lines=True)

# Combine the dataframes
combined_data = pd.concat([data1, data2], ignore_index=True)

# Display the combined data
print(combined_data.head())

combined_data.to_json('combined_data.json', orient='records')
```

```python
from sklearn.model_selection import train_test_split

# Split the combined dataset into training and testing sets
train_data, test_data = train_test_split(combined_data, test_size=0.2, random_state=42)

# If using cross-validation, further split the training set into train and validation sets
# train_data, val_data = train_test_split(train_data, test_size=0.2, random_state=42)

# Optional: Print the shapes of the resulting datasets
print("Training data shape:", train_data.shape)
print("Testing data shape:", test_data.shape)
# print("Validation data shape:", val_data.shape)
```

```python
import spacy
from spacy.training import Example
from tqdm import tqdm

# Load the spaCy English model
nlp = spacy.blank("en")

# Add named entity labels to the pipeline
ner = nlp.add_pipe("ner")

# Add your custom entity labels
ner.add_label("ATTRIBUTE")
ner.add_label("TYPE")
ner.add_label("ENTITY")
ner.add_label("OBJECT")
ner.add_label("LOC-ROOM")
ner.add_label("LOC-LEVEL")

# Prepare training data
train_examples = []
for idx, row in tqdm(train_data.iterrows(), total=len(train_data)):
    text = row["text"]
    entities = []
    for label in row["labels"]:
        start, end, entity_type = label
        entities.append((start, end, entity_type))
    train_examples.append(Example.from_dict(nlp.make_doc(text), {"entities": entities}))

# Initialize the optimizer
optimizer = nlp.begin_training()

# Train the NER model
for _ in tqdm(range(10)):  # Adjust the number of epochs as needed
    losses = {}
    for example in train_examples:
        nlp.update([example], losses=losses, drop=0.1)
    print("Losses:", losses)

# Save the trained model
nlp.to_disk("ner_model")
```

```python
import spacy

# Load the trained NER model
model_path = "/content/ner_model"  # Replace with the path to your trained model
nlp = spacy.load(model_path)

# Assuming `test_data` is a DataFrame containing your testing data
testing_data = test_data

# Evaluate the model on the testing dataset
correct_entities = 0
predicted_entities = 0
actual_entities = 0

for index, row in testing_data.iterrows():
    text = row["text"]
    entities = row["labels"]

    doc = nlp(text)
    predicted = [(ent.start_char, ent.end_char, ent.label_) for ent in doc.ents]

    # Convert predicted entities to match the format of actual entities
    predicted = [[pred[0], pred[1], pred[2]] for pred in predicted]

    # Compare predicted entities with ground truth annotations


    # Check for correct entities
    for entity in entities:
        if entity in predicted:
            correct_entities += 1
    predicted_entities += len(predicted)
    actual_entities += len(entities)

# Calculate precision, recall, and F1-score
precision = correct_entities / predicted_entities if predicted_entities > 0 else 0
recall = correct_entities / actual_entities if actual_entities > 0 else 0
f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

# Print evaluation metrics
print("Evaluation Metrics:")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-score: {f1_score:.2f}")
print(correct_entities)
```

```python
import ifcopenshell
import spacy

# Load the IFC file
ifc_file_path = "/content/Clinic_A_20110906.ifc"
ifc_file = ifcopenshell.open(ifc_file_path)

# Get all entities in the IFC file
entities = ifc_file.by_type("IfcProduct")

# Load the NER model
ner_model_path = "/content/ner_model"
ner_model = spacy.load(ner_model_path)

# Function to extract attributes using the NER model
def extract_attributes(text):
    # Tokenize the text
    doc = ner_model(text)

    # Extract named entities
    named_entities = {ent.label_: ent.text for ent in doc.ents}

    return named_entities

# Iterate over each entity
for entity in entities:
    print("Entity:")
    print(entity)
    print("Extracted Attributes:")

    # Extract relevant attributes using the NER model
    entity_text = entity.Name if hasattr(entity, "Name") and entity.Name else ""
    if entity_text:
        attributes = extract_attributes(entity_text)

        # Print the extracted attributes
        for attribute, value in attributes.items():
            print(f"{attribute}: {value}")
    else:
        print("No name attribute found or the name attribute is empty for the entity.")
```
